```python
#importing necessary packages
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import confusion_matrix,classification_report

```

```python
data=pd.read_csv('/content/gdrive/MyDrive/MachineLearning/SVM/UnivBank.csv', na_values=['?','#'])
```

```python
#Train and test split
X_train,X_test,y_train,y_test= train_test_split(data.loc[:,data.columns!='Personal Loan'], data['Personal Loan'], test_size=0.3, random_state=100)
```

```python
#Separating the numerical and categorical columns 
catcols=['Securities Account','CD Account','Education','Online','CreditCard']
waste=['ID','ZIP Code']
X_train=X_train.drop(waste, axis=1)
X_test=X_test.drop(waste,axis=1)
X_train[catcols]=X_train[catcols].astype('category')
X_test[catcols]=X_test[catcols].astype('category')
X_train.dtypes
X_test.shape
X_train_cat=X_train[catcols]
X_train_num=X_train.drop(catcols,axis=1)
X_test_cat=X_test[catcols]
X_test_num=X_test.drop(catcols,axis=1)
X_train_cat.head()
```

```python
X_train_num.corr()
```

```python
#Removing the most correlated variable
X_train_num=X_train_num.drop(['Age'],axis=1)
X_test_num=X_test_num.drop(['Age'],axis=1)
```

```python
#Imputing null values
impnum=SimpleImputer(strategy='mean')
impcat=SimpleImputer(strategy='most_frequent')
```

```python
#Imputing the null values in both categorical and numerical columns
X_train_num=pd.DataFrame(impnum.fit_transform(X_train_num),columns=X_train_num.columns)
X_test_num=pd.DataFrame(impnum.transform(X_test_num),columns=X_test_num.columns)
X_train_cat=pd.DataFrame(impcat.fit_transform(X_train_cat),columns=X_train_cat.columns)
X_test_cat=pd.DataFrame(impcat.transform(X_test_cat),columns=X_test_cat.columns)
# X_train_cat.astype('category')
# X_test_cat.astype('category')
```

```python
X_train_num.isnull().sum()
```

```python
# Standardizing the numerical data
std=StandardScaler()
X_train_num=pd.DataFrame(std.fit_transform(X_train_num), columns=X_train_num.columns)
X_test_num=pd.DataFrame(std.transform(X_test_num),columns=X_test_num.columns)
```

```python
#Encoding the categorical columns
X_train_cat=pd.get_dummies(columns=X_train_cat.columns,data=X_train_cat, drop_first=True)
X_train_cat.head()
X_test_cat=pd.get_dummies(columns=X_test_cat.columns,data=X_test_cat, drop_first=True)
```

```python
#Joining  the numerical and categorical 
Xtrain=pd.concat([X_train_num,X_train_cat], axis=1)
Xtest=pd.concat([X_test_num,X_test_cat],axis=1)
Xtest.shape
# Xtrain.shape
```

```python
#Estimating the best parameters using GridSearch
params={'penalty':['l2','l1','elasticnet','none'],'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
search=GridSearchCV(LogisticRegressionCV(),params, cv=10)
search.fit(Xtrain,y_train)
print(search.best_params_)
```

```python
#Fitting the model
model=LogisticRegressionCV(penalty='l2',cv=10,solver='newton-cg',class_weight=10)
log=model.fit(Xtrain, y_train)
```

```python
#Predicting the target variable
logtrain=model.predict(Xtrain)
logtest=model.predict(Xtest)
```

```python
#Printing the Confusion Matrix
print(confusion_matrix(logtrain,y_train))
print(confusion_matrix(logtest,y_test))
```

```python
#Printing the classification report
print(classification_report(logtrain,y_train))
print(classification_report(logtest,y_test))
```
